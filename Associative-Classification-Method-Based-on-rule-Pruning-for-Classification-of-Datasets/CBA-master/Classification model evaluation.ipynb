# Selecting 5 datasets from KDD paper (and 2 additional)

```python
def getDatasetMetrics(path):
    results = []
    data , attributes , value_type = process_data(path)
    pre_processed_data = pre_processing.pre_process(data, attributes, value_type)
    
    processed_dataFrame = pd.DataFrame.from_records(pre_processed_data)
    X = processed_dataFrame.iloc[:, :-1]
    y = processed_dataFrame.iloc[:, -1:]
    
    results.append(decision_tree_classifier(X,y.values.ravel(),cv=8).mean())
    results.append(random_forest_classifier(X,y.values.ravel(),cv=8).mean())
    results.append(support_vector_machine_classifier(X,y.values.ravel(),cv=8).mean())

    return results
```

```python
datasets = ['iris', 'zoo', 'tic-tac-toe', 'german']
datasets += ['vertebra', 'transfusion', 'horse']
resultDict = dict()

for dataset in datasets:
    res = getDatasetMetrics(dataset)
    resultDict[dataset] = res
print(resultDict)
```

```python
AccuracyResults = pd.DataFrame.from_dict(resultDict, orient='index', columns=['Decision Tree', 'Random forest', 'SVM'])
```
